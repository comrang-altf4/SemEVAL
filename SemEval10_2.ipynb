```python
reference_df = pd.read_csv('./EN/subtask-2-annotations.txt', sep='\t', header=None)
reference_df[1] = reference_df[1].str.lower()
reference_df[2] = reference_df[2].str.lower()
```

```python
# find the corresponding narrative id of each narrative in reference_df by matching the narrative name in df_n
reference_df['Narrative ID'] = None
for i in range(len(reference_df)):
    narrative_name = reference_df.iloc[i, 1]
    narrative_name = narrative_name.split(';')
    ids =[]
    for name in narrative_name:
        if 'other' == name:
            ids.append(-1)
            continue
        for j in range(len(df_n)):
            if df_n.iloc[j, 0] in name:
                ids.append(j)
    reference_df['Narrative ID'].loc[i] = ids
```

```python
reference_df['Sub_Narrative ID'] = None
for i in range(len(reference_df)):
    narrative_name = reference_df.iloc[i, 2]
    narrative_name = narrative_name.split(';')
    ids =[]
    for name in narrative_name:
        if 'other' == name:
            ids.append(-1)
            continue
        for j in range(len(df_s)):
            if df_s.iloc[j, 0] in name:
                ids.append(j)
    reference_df['Sub_Narrative ID'].loc[i] = ids
reference_df.head()
```

```python
#load article data, each article is in a separate normal text file
import os
import re

def load_article(file_path):
    with open(file_path, 'r',encoding="utf8") as file:
        data = file.read().replace('\n', '')
    return data

article_dir = './EN/raw-documents/'
article_files = os.listdir(article_dir)
article_files = sorted(article_files)

article_data = {}
for file in article_files:
    article_data[file] = load_article(article_dir + file)
```

```python
df_n = df_n.fillna('None')
df_s = df_s.fillna('None')
```

```python
#sample 80% of the data for training and 20% for testing
from sklearn.model_selection import train_test_split
train_ref, test_ref = train_test_split(reference_df, test_size=0.01, random_state=42)
```

```python
all_pairs = []
all_narrative_id = df_s['Narrative ID'].unique()
all_sub_narrative_id = df_s.index
for idx,row in tqdm(train_ref.iterrows(), total=len(train_ref)):
    article1 = article_data[row[0]]
    for i in all_narrative_id:
        article2 = f"Narrative type: {df_n.iloc[i, 0]}. Narrative description: {df_n.iloc[i, 1]}. Instruction for annotators: {df_n.iloc[i, 2]}."
        label = 1 if i in row['Narrative ID'] else -1 
        all_pairs.append((article1, article2, label))
    for i in all_sub_narrative_id:
        article2 = f"Sub-Narrative type: {df_s.iloc[i, 0]}. Sub-Narrative description: {df_s.iloc[i, 1]}. Instruction for annotators: {df_s.iloc[i, 2]}."
        label = 1 if i in row['Sub_Narrative ID'] else -1 
        all_pairs.append((article1, article2, label))
```

```python
positive_pairs = [pair for pair in all_pairs if pair[2] == 1]
negative_pairs = [pair for pair in all_pairs if pair[2] == -1]
all_pairs = negative_pairs
```

# Train model

```python
from sentence_transformers import SentenceTransformer,  losses, InputExample, util
from torch.utils.data import DataLoader, Dataset
import torch
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score

# 1. Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')
# model = SentenceTransformer('all-mpnet-base-v2')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
print(device)
```

```python
from datasets import Dataset
# 7. Train-test split
dataset_articles = [pair[0] for pair in all_pairs]
dataset_narratives = [pair[1] for pair in all_pairs]
dataset_labels = [pair[2] for pair in all_pairs]
dataset = Dataset.from_dict({
    "articles": dataset_articles,
    "narratives": dataset_narratives,
    "label": dataset_labels
})
```

```python
from sentence_transformers import (
    SentenceTransformer,
    SentenceTransformerTrainer,
    SentenceTransformerTrainingArguments,
    SentenceTransformerModelCardData,
)
from sentence_transformers.losses import MultipleNegativesRankingLoss
from sentence_transformers.training_args import BatchSamplers
from sentence_transformers.evaluation import TripletEvaluator

args = SentenceTransformerTrainingArguments(
    # Required parameter:
    output_dir="models/mpnet-base-all-nli-triplet",
    # Optional training parameters:
    num_train_epochs=3,
    per_device_train_batch_size=64,
    per_device_eval_batch_size=8,
    learning_rate=2e-5,
    warmup_ratio=0.1,
    fp16=True,  # Set to False if you get an error that your GPU can't run on FP16
    bf16=False,  # Set to True if you have a GPU that supports BF16
    # batch_sampler=BatchSamplers.NO_DUPLICATES,  # MultipleNegativesRankingLoss benefits from no duplicate samples in a batch
    batch_sampler=BatchSamplers.BATCH_SAMPLER,  # MultipleNegativesRankingLoss benefits from no duplicate samples in a batch
    # Optional tracking/debugging parameters:
    eval_strategy="no",
    # eval_steps=100,
    # save_strategy="steps",
    # save_steps=100,
    # save_total_limit=2,
    logging_steps=100,
    # run_name="mpnet-base-all-nli-triplet",  # Will be used in W&B if `wandb` is installed
)
# train_loss = losses.ContrastiveLoss(model)
```

```python
train_loss = losses.CosineSimilarityLoss(model)

trainer = SentenceTransformerTrainer(
    model=model,
    train_dataset=dataset,
    loss=train_loss,
    args=args,
)
trainer.train()
```

```python
sub_narrative_full = []
for i in all_sub_narrative_id:
        sub_narrative_full.append(f"Sub-Narrative type: {df_s.iloc[i, 0]}. Sub-Narrative description: {df_s.iloc[i, 1]}. Instruction for annotators: {df_s.iloc[i, 2]}.")
sub_narrative_full = model.encode(sub_narrative_full)
narrative_full = []
for i in all_narrative_id:
        narrative_full.append(f"Narrative type: {df_n.iloc[i, 0]}. Narrative description: {df_n.iloc[i, 1]}. Instruction for annotators: {df_n.iloc[i, 2]}.")
narrative_full = model.encode(narrative_full)
```

```python
# 8. Evaluate the model
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics.pairwise import cosine_similarity

# for each article, find narrative and sub-narrative that has similarity score > 0.5
pred_narrative = []
pred_sub_narrative = []
for i in range(len(embeddings)):
    sim_narrative = cosine_similarity([embeddings[i]], narrative_full)
    sim_sub_narrative = cosine_similarity([embeddings[i]], sub_narrative_full)
    # narrative with score > 0.5
    pred_narrative.append(np.where(sim_narrative > 0.7)[1].tolist())
    # sub-narrative with score > 0.5
    pred_sub_narrative.append(np.where(sim_sub_narrative > 0.7)[1].tolist())
```

```python
res = []
for i in range(len(pred_sub_narrative)):
    res_narrative = []
    res_sub_narrative = []
    check_narrative = {k:False for k in pred_narrative[i]}
    for j in pred_sub_narrative[i]:
        narrative_id = df_s.iloc[j, 3]
        if narrative_id in pred_narrative[i]:
            res_narrative.append(narrative_id)
            res_sub_narrative.append(j)
            check_narrative[narrative_id] = True
    for k,v in check_narrative.items():
        if not v:
            res_narrative.append(k)
            res_sub_narrative.append(-1)
    if len(res_narrative) == 0:
        res_narrative.append(-1)
        res_sub_narrative.append(-1)
    res.append((res_narrative, res_sub_narrative))
```
